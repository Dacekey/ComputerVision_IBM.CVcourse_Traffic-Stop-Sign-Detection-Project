# **Manipulating Image**

## Copying Image
Understand:
- Different Memory Address
- If change the original image, the variable is oriented will also be change




```python
from PIL import Image
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Using this to check the correct directory if meeting error
# os.path.exists("Directory_of_my_image")

direc = os.getcwd()
iname = "image_1.png"
image = f"{direc}/Image_Storage/{iname}"

# Two code lines below is the same
robin = cv2.imread(image)
robin = cv2.cvtColor(robin, cv2.COLOR_BGR2RGB) # Convert BGR -> RGB
# robin = np.array(Image.open(image)) # RGB

robin_pil = Image.open(image)

id(robin)
print("The id of Robin is:",id(robin))
A = robin
print("The id of A is:",id(A))
B = robin.copy()
print("The id of B is:",id(B))

# =====

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(robin)
axes[0].set_title("robin")
axes[0].axis("off")

axes[1].imshow(A)
axes[1].set_title("A")
axes[1].axis("off")

axes[2].imshow(B)
axes[2].set_title("B")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```

```python
robin[:,:,] = 0

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(robin)
axes[0].set_title("robin")
axes[0].axis("off")

axes[1].imshow(A)
axes[1].set_title("A")
axes[1].axis("off")

axes[2].imshow(B)
axes[2].set_title("B")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```

## Flipping Image




### **Pillow (PIL)**

```python
from PIL import ImageOps
```

```python
robin_flip_pil = ImageOps.flip(robin_pil)

fig, axes = plt.subplots(1, 1, figsize=(2, 2))

axes.imshow(robin_flip_pil)
axes.set_title("robin_flip_pil")
axes.axis("off")

plt.tight_layout()
plt.show()
```

```python
robin_mirror_pil = ImageOps.mirror(robin_pil)

fig, axes = plt.subplots(1, 1, figsize=(2, 2))

axes.imshow(robin_mirror_pil)
axes.set_title("robin_mirror_pil")
axes.axis("off")

plt.tight_layout()
plt.show()
```

```python
robin_ftb_pil = robin_pil.transpose(Image.Transpose.FLIP_TOP_BOTTOM)

fig, axes = plt.subplots(1, 1, figsize=(2, 2))

axes.imshow(robin_ftb_pil)
axes.set_title("robin_transpose_Top-Bottom")
axes.axis("off")

plt.tight_layout()
plt.show()
```

### **OpenCV**

```python
robin_flip_cv2 = cv2.flip(robin,0)
# cv2.flip(a, b): a -> src, b -> flipcode
# flipcode domain: [-1, 0, 1] : [Both axes, Vertical flip, Horizontal flip]

fig, axes = plt.subplots(1, 1, figsize=(2, 2))

axes.imshow(robin_flip_cv2)
axes.set_title("robin_flip_vertical")
axes.axis("off")

plt.tight_layout()
plt.show()
```

#### Rotate function

```python
robin_rotate_cv2 = cv2.rotate(robin, cv2.ROTATE_90_CLOCKWISE)

fig, axes = plt.subplots(1, 1, figsize=(2, 2))

axes.imshow(robin_rotate_cv2)
axes.set_title("robin_rotate_90degree")
axes.axis("off")

plt.tight_layout()
plt.show()
```

# **Summarise**
- Copying Image
  + Memory Address
  + Changes from the origin are referenced to assigned variable
- Flipping Image
  + Pillow (pil):
    * Flip()

  + OpenCV (cv2):
