# **Mini-Batch Gradient Descent**

### Allows train model with more data
### Previous (S2) we use every samples for Gradient Descent (Batch Gradient Descent | BGD)
### Now (S3) with MGD, we use a few samples at a time for each iteration.

# **Notation:**
#### **Epoch:** the time of being used all the samples in the dataset
#### **Batch Size:** the number of samples per iteration
#### **Batch Gradient Descent:** where one iteration equals one epoch
#### **Total Loss:** is the sum (or sometimes the average) of the individual losses of multiple samples, typically within a mini-batch, used to compute the gradient during training.

# How to determine the number of iterations for different batch sizes and epochs

### (@) Calculate the total loss before each iteration


### (@) At the end of each epoch we calculate the accuracy on the validation data.
### (@) We repeat the process for the next iteration
### (@) If the accuracy decreases we have trained too much. => This is called **Overfitting**
