# **Project Overview**

## **I./ Project Overview**

You've learned to train and apply custom image classifiers in this course. In the final project, you'll practice applying skills to create a custom classifier that identifies **traffic signs** i.e., whether or not an image contains a **stop sign**. Practical machine learning skills are highlighted in this project: **image data processing, feature extraction, model training, and model accuracy assessment**. The objective is to develop an accurate image classifier based on the given dataset to **recognize stop signs**.

## **II./ Project Scenario**

You have just been recruited as a Junior Data Scientist at Jokwu, a self-driving car company based in Cape Town, South Africa. The firm has created the hardware components of the vehicle and is currently developing vision-based systems to recognize road signs. As part of this work, you have been instructed to develop a model to recognize stop signs in photographs. You will use a JupyterLab-based environment in the Final assignment: **Building an Image Classifier with Transfer Learning** to train and build the model. The model will be a foundation for subsequent safety and motion control features. Your assignment is to train and evaluate a model that predicts whether an image includes a stop sign using a labeled dataset, your team has provided.
